```python
import pandas as pd
import numpy as np
import os.path
import matplotlib.pyplot as plt
```

```python
snaps = pd.read_csv('snapscore.txt',skiprows=0,sep=',',header=0)
snaps['difference'] = snaps['new'] -snaps['old']
```

```python
df = pd.read_csv('snapscore.txt',skiprows=0,sep=',',header=0, usecols=['new'], nrows=len(snaps.index))
df50 = df.tail(50)
df.plot()
plt.show()
print("LAST 50 UPDATES")
df50.plot()
plt.show()
#histgram = snaps.drop(['check','old',"new"],axis=1)
#histgram.hist(bins=20, linewidth=1, edgecolor="white")
#plt.show()
```

```python
#Find how many updates the snapscore grew by 30
snapsjump = snaps[(snaps.difference>=30)]
#print(snapsjump)
snapsjumpcount = snapsjump.difference.value_counts()
print("____________________________________________________________")
print(len(snapsjump.index),"snapscore updates which jumped more than 30, out of",len(snaps.index),"updates")
```

```python
a = input("FOR HOW MANY OF THE LAST UPDATES DO YOU WANT TO COMPARE")
b = int(a)

snapsDifference = snaps.drop(['check','old',"new"],axis=1)
snapsDifference.tail(50).plot()
plt.show()
df.tail(50).plot()
plt.show()
```

```python
#Find average snapscore growth
averageSnapsDifference = snapsDifference["difference"].mean()
roundedAverage = round(averageSnapsDifference, 3)
print("Snapscore grows an average of",'\033[1m' + str(roundedAverage) + '\033[0m',"each update")

print("Highest snapscore update was",'\033[1m' + str(snapsDifference["difference"].max()) + '\033[0m')
```

```python
#Sort differences
pd.set_option('max_rows', None)
snapsDifference.sort_values(by=['difference'], inplace=True)
print(snapsDifference)

#usualdiff = snaps.difference.value_counts()
#print(usualdiff)
```

```python
#Print the top 20 greatest snapscore updates 
snapsDifference.sort_values(by=['difference'], ascending=False, inplace=True)
snapsDifference.drop(snapsDifference.tail(len(snapsDifference)-20).index, inplace = True)
print(snapsDifference)
print("Most recent check was update",'\033[1m' + str(len(snaps)-1) + '\033[0m')
```

```python
#Find how many of the greatest jumps took place within the last 20 updates 
top20Count = 0
snapDifferencewithCheck = snaps.sort_values(by=['difference'], ascending=False)
snapDifferencewithCheck.drop(snapDifferencewithCheck.tail(len(snapDifferencewithCheck)-20).index, inplace = True)

for index, row in snapDifferencewithCheck.iterrows():
    if int(row['check']) >= (len(snaps)-1)-20:
        print("update",row['check'],"had a difference of",row['difference'])
        top20Count = top20Count + 1
print("__________________________________________________________")
print("Out of",'\033[1m' + str(len(snaps)-1) + '\033[0m',"updates...")
print('\033[1m' + str(top20Count) + '\033[0m',"of the biggest updates took place within the last 20 updates")
```
